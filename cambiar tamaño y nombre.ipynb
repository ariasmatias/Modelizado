```python
from PIL import Image, UnidentifiedImageError
import os
import re

# Define the root folder containing the images
root_folder = 'C:/Users/diego/Desktop/TSCDIA/Modelizado de Sistemas de IA/TP1/Modelizado/image_set'
new_size = (256, 256)  # Desired size (width, height)

# Function to extract the current number from the filename
def extract_number(filename):
    match = re.search(r'_(\d+)$', filename)  # Look for a number after the last underscore
    return int(match.group(1)) if match else None

# Function to find the highest number in the current subfolder
def get_highest_number_in_folder(folder_path):
    highest_number = 0
    for filename in os.listdir(folder_path):
        num = extract_number(filename)
        if num is not None and num > highest_number:
            highest_number = num
    return highest_number

# Counters for modified images
resized_count = 0
renamed_count = 0

# List to track modified image names
modified_images = []

# Traverse all subdirectories and files
for dirpath, dirnames, filenames in os.walk(root_folder):
    # Get the highest number in the current folder
    highest_number = get_highest_number_in_folder(dirpath)  
    new_image_number = highest_number + 1  # Start numbering new images from the next number

    for filename in filenames:
        img_path = os.path.join(dirpath, filename)

        if filename.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.gif')):  # Supported extensions
            try:
                # Open the image and get the current size
                img = Image.open(img_path)
                current_size = img.size
                is_correct_size = current_size == new_size
                
                # Determine folder and subfolder names for naming convention
                subfolder_name = os.path.basename(dirpath)
                folder_name = os.path.basename(os.path.dirname(dirpath))

                # Create the expected filename prefix without the extension
                expected_name_prefix = f"{folder_name}_{subfolder_name}_"
                # Check if the filename follows the naming convention (without the extension)
                follows_naming_convention = filename.startswith(expected_name_prefix)

                # If the image is not the correct size or does not follow the naming convention
                if not is_correct_size or not follows_naming_convention:
                    # Resize the image if necessary
                    if not is_correct_size:
                        img_resized = img.resize(new_size)
                        img.close()  # Close the image to release the file before renaming or saving

                        # Convert RGBA to RGB before saving if necessary
                        if img_resized.mode == 'RGBA':
                            img_resized = img_resized.convert('RGB')

                        img_resized.save(img_path)  # Save the resized image, overwriting the original
                        resized_count += 1
                        modified_images.append(os.path.basename(img_path))

                    # If the filename is incorrect, rename it
                    if not follows_naming_convention:
                        new_filename = f"{folder_name}_{subfolder_name}_{new_image_number}.jpg"
                        new_img_path = os.path.join(dirpath, new_filename)

                        # Check if the new file already exists
                        while os.path.exists(new_img_path):
                            new_image_number += 1
                            new_filename = f"{folder_name}_{subfolder_name}_{new_image_number}.jpg"
                            new_img_path = os.path.join(dirpath, new_filename)

                        # Ensure the file is not being used by another process
                        os.rename(img_path, new_img_path)
                        renamed_count += 1
                        modified_images.append(new_filename)

            except UnidentifiedImageError:
                print(f"Cannot identify image file: {img_path}")
            except PermissionError:
                print(f"Permission denied for file: {img_path}")
            finally:
                img.close()  # Close the image to release the file

# Summary of modifications
print(f"Total images resized: {resized_count}")
print(f"Total images renamed: {renamed_count}")
print("Modified images:")
for image in modified_images:
    print(image)
print("All images processed.")
```
